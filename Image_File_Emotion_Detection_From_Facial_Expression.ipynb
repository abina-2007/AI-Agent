```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt

def detect_emotion(image_path):
    try:
        # Analyze the image for emotions
        # 'actions' can include ['age', 'gender', 'race', 'emotion']
        results = DeepFace.analyze(img_path=image_path, actions=['emotion'])

        # DeepFace returns a list (in case there are multiple faces)
        for face in results:
            print(f"Dominant Emotion: {face['dominant_emotion']}")
            print(f"Confidence: {face['emotion'][face['dominant_emotion']]:.2f}%")

        # Optional: Display the image
        img = cv2.imread(image_path)
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img_rgb)
        plt.title(f"Detected: {results[0]['dominant_emotion']}")
        plt.axis('off')
        plt.show()

    except Exception as e:
        print(f"Error: {e}")

# Replace with your image file name
detect_emotion('emotions.jpg')
detect_emotion('my_photo.jpg')
```
